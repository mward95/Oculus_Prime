```python
#Split data into X and y
X = df.drop(columns=["MYOPIC", "DIOPTERHR","STUDYYEAR" ])
y = df["MYOPIC"]
```

```python
X.head()
```

```python
y.head()
```

Split the Data

```python
# Split the data into X_train, X_test, y_train, y_test
# X, y = make_classification(random_state=1)
# X = pd.DataFrame(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.2)
# scaler = StandardScaler().fit(X_train)
# X_train_scaled = scaler.transform(X_train)
# X_test_scaled = scaler.transform(X_test)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# model = RandomForestClassifier(random_state=1)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=0, n_estimators= 30)
model.fit(X_train, y_train)
```

### The n-estimators were iterated to find the best model score. It came out to be 30 random trees!

```python
model.score(X_test, y_test)
```

### Checking to see if scaled values give a better model score 

```python
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
model = RandomForestClassifier(random_state=0, n_estimators= 30)
model.fit(X_train_scaled, y_train)
```

```python
model.score(X_test_scaled, y_test)
```

```python


# print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
# print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

```python
y_predicted = model.predict(X_test_scaled)
```

To see how well the model performed - we'd use the confusion Matrix - y_test is our truth and y_predicted is our predictions

```python
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize = (7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.title(label = 'Random Forest - Confusion Matrix (Model Score 93.5%)', 
         fontsize = 15, 
         fontweight = 20)
```

## NOTE: run the code only once to get trees - it takes sometime to run and generate all trees and branches

```python
import os
from sklearn.tree import export_graphviz
import six
import pydot
from sklearn import tree
dotfile = six.StringIO()
i_tree = 0
for tree_in_forest in model.estimators_:
    export_graphviz(tree_in_forest,out_file='tree.dot',
    feature_names=X.columns,
    filled=True,
    rounded=True)
    (graph,) = pydot.graph_from_dot_file('tree.dot')
    name = 'tree' + str(i_tree)
    graph.write_png(name+  '.png')
    os.system('dot -Tpng tree.dot -o tree.png')
    i_tree +=1
```

### The above code will spit out all Trees in Random Forest code in your folder - Below is a Snapshot of Tree0.png

### Code below lists the  hierarchy of Important Features that affects model fitting

```python
#modelname.feature_importance_
y = model.feature_importances_
#plot
plt.figure(figsize=(10,8))
fig, ax = plt.subplots() 
width = 0.4 # the width of the bars 
ind = np.arange(len(y)) # the x locations for the groups
ax.barh(ind, y, width, color = "green")
ax.set_yticks(ind+width/10)
ax.set_yticklabels(X.columns, minor=False)
plt.title('Feature importance in RandomForest Classifier(n=30)')
plt.xlabel('Relative importance')
plt.ylabel('feature') 

fig.set_size_inches(6.5, 8.5, forward=True)
```
